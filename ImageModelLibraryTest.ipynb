```python
import torch
from torchvision import models
from torchvision import transforms
from PIL import Image

print(f"Torch Version: {torch.__version__}")
```

```python
# Image Models
dir(models)
```

```python
# Creating an instance of resnet101
# A 101 layer convolutional neural network, trained on ImageNet dataset with 1.2M images and 1000 categories.

resnet = models.resnet101(pretrained=True)  # Also download the weights of the trained model
```

```python
resnet
```

```python
# preprocess function

preprocess = transforms.Compose(
    [transforms.Resize(256),
     transforms.CenterCrop(224),
     transforms.ToTensor(),
     transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225])]
)
```

```python
img = Image.open('/content/Teutonic Knight Red.jpg')
```

```python
img
```

```python
# preprocess the image

img_t = preprocess(img)
batch_t = torch.unsqueeze(img_t, 0)
```

```python
img_t
```

```python
# Evaluation mode
resnet.eval()
```

```python
out = resnet(batch_t)
out
```

```python
# Loading the labels

with open('/content/imagenet_classes.txt') as f:
  labels = [line.strip() for line in f.readlines()]
```

```python
# Finding the index which model predicted

_, index = torch.max(out, 1)
```

```python
percentage = torch.nn.functional.softmax(out, dim=1)[0]*100
labels[index[0]], percentage[index[0]].item()
```
